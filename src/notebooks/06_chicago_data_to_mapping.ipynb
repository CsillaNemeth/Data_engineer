```python
import pandas as pd
pd.set_option("display.max_columns", 30)
import requests
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
current_datetime = datetime.now() - relativedelta(months=2)

formatted_datetime = current_datetime.strftime('%Y-%m-%d')
formatted_datetime
```

```python
# Get T-2 month's  trips data

url = f"https://data.cityofchicago.org/resource/ajtu-isnz.json?$where=trip_start_timestamp >= '{formatted_datetime}T00:00:00' AND trip_start_timestamp <= '{formatted_datetime}T23:59:59'&$limit=30000"

response = requests.get(url)
data = response.json()
taxi_trips = pd.DataFrame(data)
taxi_trips.info()
```

Transformation: Handle NaN values

```python
taxi_trips.drop(["pickup_census_tract", "dropoff_census_tract"], axis=1, inplace=True)
taxi_trips.drop(["pickup_centroid_location", "dropoff_centroid_location"], axis=1, inplace=True)

taxi_trips.dropna(inplace=True)
```

Transformation: renaming

```python
taxi_trips.rename(columns={'pickup_community_area': 'pickup_community_area_id', 'dropoff_community_area': 'dropoff_community_area_id'}, inplace=True)
```

Transformation: create helper column for weather

```python
taxi_trips['trip_start_timestamp'] = pd.to_datetime(taxi_trips['trip_start_timestamp'])
```

```python
taxi_trips['datetime_for_weather'] = taxi_trips['trip_start_timestamp'].dt.floor('h')
```

```python
# Get T-2 months' weather data.

url = 'https://archive-api.open-meteo.com/v1/era5?'

current_datetime = datetime.now() - relativedelta(months=2)
formatted_datetime = current_datetime.strftime("%Y-%m-%d")
params = {
    "latitude": 41.85,
    "longitude": -87.65,
    "start_date": formatted_datetime,
    "end_date": formatted_datetime,
    "hourly": "temperature_2m,wind_speed_10m,rain,precipitation"
}


response = requests.get(url, params=params)
weather_data = response.json()

# Create a  weather dataframe.

weather_data_filtered = {
    'datetime': weather_data['hourly']["time"],
    'tempereature': weather_data['hourly']['temperature_2m'],
    'wind_speed': weather_data['hourly']['wind_speed_10m'],
    'rain': weather_data['hourly']['rain'],
    'precipitation': weather_data['hourly']['precipitation']
}

weather_df = pd.DataFrame(weather_data_filtered)
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])

weather_df.head()

weather_df.to_csv('weather_data_date.csv',  index = False)
```

```python
taxi_trips_with_weather = taxi_trips.merge(weather_df, left_on='datetime_for_weather', right_on='datetime')
```

```python
taxi_trips_with_weather.head()
```

Transformation: Data type conversions

```python

data_types = {
    "trip_end_timestamp": "datetime64[ns]",
    "trip_seconds": "int32",
    "trip_miles": "float",
    "pickup_community_area_id": "int8",
    "dropoff_community_area_id": "int8",
    "fare": "float",
    "tips": "float",
    "tolls": "float",
    "extras": "float",
    "trip_total": "float",
}

taxi_trips = taxi_trips.astype(data_types)
```

```python
taxi_trips.info()
```

Data modeling

```python
# Create a payment type master table

payment_type_master = taxi_trips['payment_type'].drop_duplicates().reset_index(drop=True)

payment_type_master = pd.DataFrame( 
    {
    'payment_type_id': range(1, len(payment_type_master) +1),
    'payment_type': payment_type_master

}
)
payment_type_master
```

```python
# Create a company master table

company_master = taxi_trips['company'].drop_duplicates().reset_index(drop=True)

company_master = pd.DataFrame( 
    {
    'company_id': range(1, len(company_master) +1),
    'company': company_master

}
)
company_master
```

```python
payment_type_master.to_csv('payment_type_master.csv', index=False)
company_master.to_csv('company_master.csv', index=False)
```
